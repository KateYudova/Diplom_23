```python
#1
token_embeddings = torch.stack(BERT_hidden_states, dim = 0)
token_embeddings.size()
```

```python

token_embeddings = torch.squeeze(token_embeddings, dim = 1)
token_embeddings.size()
```

```python
#14 tokens, 13 layers
token_embeddings = token_embeddings.permute(1,0,2)
token_embeddings.size()
```

1 способ. Конкатинация последних 4 слоев

```python
token_vecs_cat = [] #shape [14*3,072]

#token_embeddings is [14*13*768] tensor
#Для каждого токена в предложении:

for token in token_embeddings:
  #token is [13*768]
  #cat_vec is length = 4*768
  cat_vec = torch.cat((token[-1], token[-2],token[-3],token[-4]), dim = 0)

  #Используем cat_vec чтобы представить токен
  token_vecs_cat.append(cat_vec)

print('Shape is: %d * %d' % (len(token_vecs_cat), len(token_vecs_cat[0])))
#Теперь для каждого токена (слова) у нас есть вектор 3072 размерности
```

2 способ. Сложение 4 последних слоев

```python
token_vecs_sum = [] # [14*768]

for token in token_embeddings:
  sum_vec = torch.sum(token[-4:], dim = 0)
  token_vecs_sum.append(sum_vec)

print('Shape is: %d * %d' % (len(token_vecs_sum), len(token_vecs_sum[0])))
#Теперь для каждого токена (слова) у нас есть вектор 768 размерности
```
